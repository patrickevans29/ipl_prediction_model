## Team1_MVP_average and Team2_MVP_average

```python
complete_df = complete_df.drop_duplicates(subset='ID', keep='first')
```

```python
def get_match_winner(player_of_match, team1_players, team2_players):
    if pd.isna(player_of_match):
        return 'N/A'  
    
    if pd.isna(team1_players):
        team1_players = []  
    
    if pd.isna(team2_players):
        team2_players = [] 
    
    if player_of_match in team1_players:
        return 'Team1'
    elif player_of_match in team2_players:
        return 'Team2'
    else:
        return 'N/A'
```

```python

complete_df['Team_MVP'] = complete_df.apply(lambda row: get_match_winner(row['Player_of_Match'], row['Team1Players'], row['Team2Players']), axis=1)
```

```python
complete_df.columns
```

```python
complete_df.head()
```

```python
def replace_team_mvp_with_name(row):
    if row['Team_MVP'] == 'Team1':
        return row['Team1']
    elif row['Team_MVP'] == 'Team2':
        return row['Team2']
    else:
        return row['Team_MVP']
```

```python
complete_df['Team_MVP'] = complete_df.apply(replace_team_mvp_with_name, axis=1)
```

```python
complete_df.head()
```

```python
team_mvp_counts = complete_df['Team_MVP'].value_counts().reset_index()
team_mvp_counts
```

```python
team_mvp_counts.columns = ['Team_MVP', 'MVP_Count']

# Create the columns Team1_MVP_appearances e Team2_MVP_appearances
complete_df = complete_df.merge(team_mvp_counts, left_on='Team1', right_on='Team_MVP', how='left').fillna(0)
complete_df.rename(columns={'MVP_Count': 'Team1_MVP_appearances'}, inplace=True)

complete_df = complete_df.merge(team_mvp_counts, left_on='Team2', right_on='Team_MVP', how='left').fillna(0)
complete_df.rename(columns={'MVP_Count': 'Team2_MVP_appearances'}, inplace=True)

# Drop columns
complete_df.drop(['Team_MVP_x', 'Team_MVP_y', 'Team_MVP'], axis=1, inplace=True)
```

```python
complete_df.head()
```

```python
# Now, let's create the average of MVP appearences
total_games_team1 = complete_df['Team1'].value_counts()
total_games_team2 = complete_df['Team2'].value_counts()
```

```python
total_games = total_games_team2 + total_games_team1
```

```python
total_games.sort_values(ascending=False)
```

```python
complete_df['Team1_MVP_average'] = complete_df['Team1_MVP_appearances'] / total_games[complete_df['Team1']].values
complete_df['Team2_MVP_average'] = complete_df['Team2_MVP_appearances'] / total_games[complete_df['Team2']].values
```
